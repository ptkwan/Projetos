## O que √© o KNN? ü§î <a name="knn"></a>

<br>

O KNN (K-nearest neighbors, ou ‚ÄúK-vizinhos mais pr√≥ximos‚Äù) costuma ser um dos primeiros algoritmos aprendidos por iniciantes no mundo do aprendizado de m√°quina. Em resumo, o KNN tenta classificar cada amostra de um conjunto de dados avaliando sua dist√¢ncia em rela√ß√£o aos vizinhos mais pr√≥ximos. Se os vizinhos mais pr√≥ximos forem majoritariamente de uma classe, a amostra em quest√£o ser√° classificada nesta categoria.

<br>

Ou seja, escolhemos um n√∫mero **"K"** de "vizinhos pr√≥ximos" para definirmos a classifica√ß√£o do item que n√£o est√° classificado. E para esse projeto decidimos definir o nosso valor de **"K"** igual a 5

## 2. Criando a fun√ß√£o `distancia_euclidiana` que verifica a distancia entre 2 pontos.<br> <a name="2"></a>
>Esse √© o cora√ß√£o do projeto, a nossa ideia foi criar n√£o s√≥ uma fun√ß√£o para ser utilizada nesse conjunto de dados espec√≠ficos, mas sim uma que funcionaria para calcular a dist√¢ncia com quaisquer quantidades de "dimens√µes"

````python
def distancia_euclidiana(ponto1, ponto2):
    somatoria = 0
    for i in range(len(ponto1)):
        somatoria += (ponto1[i] - ponto2[i]) ** 2
    return somatoria ** 0.5
````

## 3. Como decidir um empate?

>A fun√ß√£o `desvioPadrao` foi criada para que, se ouvesse dois pontos empatados em rela√ß√£o a dist√¢ncia, decidir√≠amos a menor dist√¢ncia pelo desvio padr√£o. <a name="3"></a>

````python
def calcular_desvio_padrao(lista):
  # Calcular a m√©dia
  media = (lista[0] + lista[1]) / 2

  # Calcular os desvios
  desvio1 = (lista[0] - media)
  desvio2 = (lista[1] - media)

  # Calcular a vari√¢ncia
  variancia = (desvio1 ** 2 + desvio2 ** 2) / 2

  # Calcula o desvio padr√£o
  desvio_padrao = (variancia) ** 0.5

  # Retorna o resultado
  return desvio_padrao
````

## 4. Calculando as dist√¢ncias entres os pontos

> A Fun√ß√£o `matriz_de_distancias` realiza o c√°lculo de dist√¢ncia euclidiana entre cada um dos 30 CPF's que precisam ser classificados e os 120 j√° classificados, retornando uma matriz 2x2 relacionando a dist√¢ncia entre os pontos de `no_class` e `data` <a name="4"></a>

````python
def matriz_de_distancias(data,no_class):

  # Armazena apenas as classes dos itens de "data"
  lista_de_classes = []

  # Armazena a dist√¢ncia de um ponto com o outro ("data" e "no_class")
  distancias_de_um_ponto = []

  # Armazena a dist√¢ncia de todos os pontos ("data" e "no_class")
  distancias_todos_pontos = []

  # Vari√°veis tempor√°rias para tratar e relacionar as classes com as dist√¢ncias
  zip_distancias_classes = []
  distancias_classes_lista = []
  lista_distancia = []

  # Retorna a dist√¢ncia e a classe em tuplas
  distancias_classificadas = []

  # Preenchendo a lista com as classes dos itens de "data"
  for i in range(len(data)):
      lista_de_classes.append(data[i][1])

  # Devolve uma Matriz 2x2 relacionando a dist√¢ncia entre os pontos de "no_class" e "data" com seus respectivos √≠ndices
  for i in range(len(no_class)):
    for j in range(len(data)):
      distancias_de_um_ponto.append(distancia_euclidiana(no_class[i][2], data[j][2]))

    # Aqui armazenamos a dist√¢ncia de cada um dos pontos em uma lista que cont√©m todos os pontos
    distancias_todos_pontos.append(distancias_de_um_ponto)

    # Aqui zeramos a vari√°vel para pegar o pr√≥ximo ponto e n√£o misturar com os outros, depois salvar novamente
    distancias_de_um_ponto = []

    # Aqui juntamos todas as dist√¢ncias com suas respectivas classes, devolvendo em tupla
    zip_distancias_classes = list(zip(distancias_todos_pontos[i],lista_de_classes))
    distancias_classificadas.append(zip_distancias_classes)

    # Aqui transformamos a distancias_classificadas em lista para podermos trabalhar com os valores
    [lista_distancia.append(list(distancias_classificadas[i][k])) for k in range(len(data))]

    distancias_classes_lista.append(lista_distancia)

    # Limpa a lista para reiniciar o processo de itera√ß√£o novamente
    lista_distancia = []

  return distancias_classes_lista
````

##5. Classificando os CPFs se baseando nas dist√¢ncias calculadas

> A fun√ß√£o `contagem_de_classe` ir√° identificar qual a classe de cada CPF n√£o classificado baseado nas **"K"** dist√¢ncias encontradas (como dito antes, definimos K como 5) <a name="5"></a>

````python
def contagem_de_classe(lista):

    pro = 0 # "Promotor"
    neu = 0 # "Neutro"
    det = 0 # "Detrator"

    lista_temp = lista.copy()

    # Organiza a lista e seleciona os 5 primeiros
    lista_temp.sort()
    lista_k_menores = lista_temp[:5] # Aqui pegamos os 5 primeiros (valor que definimos para o "K")

    classe_definida = ""

    # Fazendo a contagem de todas as ocorr√™ncias das classes
    for _, palavra in lista_k_menores:
        if palavra == 'Promotor':
            pro += 1
        elif palavra == 'Neutro':
            neu += 1
        elif palavra == 'Detrator':
            det += 1

    # Cria uma lista determinando a quantidade de cada ocorr√™ncia de classe
    qtde_classes = [pro, neu, det]
    cont_classes = []

    # Verifica o que mais apareceu e armazena na "qtde_classes"
    for i in range(len(qtde_classes)):
        maior = max(qtde_classes)
        # Conta quantas ocorr√™ncias houveram do maior item
        ocorrencia_maior = qtde_classes.count(maior)

        # Combina os itens da "qtde_classes" com as respectivas classes
        cont_classes = list(zip(qtde_classes,["Promotor","Neutro","Detrator"]))

        # Ordena a lista da classe que mais apareceu para a que menos apareceu
        cont_classes.sort(reverse = True)


    # Caso haja mais que uma ocorr√™ncia, houve empate, portanto ele executa o "If"
    if ocorrencia_maior > 1:
        classe_1 = cont_classes[0][1]
        classe_2 = cont_classes[1][1]

        print(f"\nHouve um empate entre {classe_1} e {classe_2}")

        # Armazena os valores de dist√¢ncia de cada uma das classes que empataram
        distancias_1 = []
        distancias_2 = []

        # Localizar quais as dist√¢ncias das classes que empataram para calcular o desvio padr√£o
        for i in range(len(lista_k_menores)):
            if classe_1 == lista_k_menores[i][1]:
                distancias_1.append(lista_k_menores[i][0])
            elif classe_2 == lista_k_menores[i][1]:
                distancias_2.append(lista_k_menores[i][0])

        # Calcula o desvio padr√£o
        desvio1 = calcular_desvio_padrao(distancias_1)
        desvio2 = calcular_desvio_padrao(distancias_2)

        print(f"\nO desvio padrao da {classe_1} √© = {desvio1}")
        print(f"\nO desvio padrao da {classe_2} √© = {desvio2}")

        if desvio1 < desvio2:
            classe_definida = classe_1
            print(f"\nComo o desvio padrao da {classe_1} √© menor, temos a menor dispersao dos dados, portanto classificamos como {classe_1}")
            return classe_definida

        elif desvio2 < desvio1:
            classe_definida = classe_2
            print(f"\nComo o desvio padrao da {classe_2} √© menor, temos a menor dispersao dos dados, portanto classificamos como {classe_2}")
            return classe_definida

        else:
            print("\nOs desvios s√£o iguais, portanto podemos fazer um random Tie Break")
            classe_definida = classe_1
            return classe_definida

    # Se nao for empate, ele vai simplesmente salvar o primeiro valor
    else:
        classe_definida = cont_classes[0][1]
        return classe_definida
````
